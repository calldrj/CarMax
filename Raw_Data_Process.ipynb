```python
import pandas as pd
import numpy as np
import re
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('CaseCompetitionData2021.csv')
df
```

```python
df.info()
```

```python
df['purchase_make'] = df['purchase_make'].apply(lambda x:  'MBZ' if x == 'MERCEDES-BENZ' else x[:4])

df['purchase_vehicle_year'] = df['purchase_vehicle_year'].apply(lambda x: 2014 - x)

df.at[27160, 'purchase_price'] = '70001 - 75000'
df.at[159987, 'purchase_price'] = '70001 - 75000'
df.drop(index=[283160], inplace=True)
df.reset_index(drop=True, inplace=True)
df['purchase_price'] = df['purchase_price'].apply(lambda x: int(x[-5:-3]))

age = lambda x: 0 if x == '?' else 100 if x in ['101+', '91 - 100'] else int(x[-2:])
df['customer_age'] = df['customer_age'].apply(age)
a = df[df['customer_age'] != 0]['customer_age'].value_counts(normalize=True)
b = a.values
k = a.index
e = lambda x : np.random.choice(k, p=b) if x == 0 else x
df['customer_age'] = df['customer_age'].apply(e)

df['customer_income'] = df['customer_income'].apply(lambda x: x.replace('?', ' 0').replace('200001+', ' 220000'))
income = r' (\d+)'
f = lambda x: int(re.search(income, x).group().replace(' ', '')) / 1000
df['customer_income'] = df['customer_income'].apply(f)

d = df[df['customer_distance_to_dealer'] != '?'].customer_distance_to_dealer.value_counts(normalize=True)
s = [int(x) for x in d.index] 
y = d.values
z = lambda x : np.random.choice(s, p=y) if x == '?' else int(x)
df['customer_distance_to_dealer'] = df['customer_distance_to_dealer'].apply(z)

t = df[df.post_purchase_satisfaction != '?'].post_purchase_satisfaction.value_counts(normalize=True)
v =  t.values
i = [int(x) for x in t.index]
h = lambda x : np.random.choice(i, p=v) if x == '?' else int(x)
df['post_purchase_satisfaction'] = df['post_purchase_satisfaction'].apply(h)
```

```python
df.info()
```

```python
df.drop(columns=['insert_num'], inplace=True)
df.columns = ['Make', 'Model', 'Years', 'Price', 'Trade_In', 
              'Financed', 'Age', 'Income', 'Gender', 'Returned', 
              'Distance', 'Rate', 'Warranty', 'Multiple_Buys']
```
